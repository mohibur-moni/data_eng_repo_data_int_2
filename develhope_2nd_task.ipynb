```python
import pandas as pd
import matplotlib.pyplot as plt

paly_store = pd.read_csv('https://download1589.mediafire.com/u1ro6t7vkkrg/6zljdqj08o2sm30/googleplaystore.csv')




df = paly_store[paly_store['Price'] != 'Everyone']
```

```python
# working with installs
df_installs = df['Installs'].str.replace(r'\W', '') # remove special chracter to make it float using regex

""" r'/ means start regex, smaller case w means without special charchter but we want those intact but special charachter.
For that we have used upper case W, and finally replace with nothing. """


df_installs = pd.to_numeric(df_installs, errors='coerce') # converting data type to numeric and raplce nan if conversation reaise an error
df_installs.isna().sum() # checking how many nan values we still have

df['Installs'] = df_installs # repalce new column values with old one 

#df.sort_values(by='Installs',ascending=False)[0:3] # top three application based on installs.


# working with price
df_price = df['Price']
df_price.unique() # cheking unique values

df_price = df['Price'].str.replace(r'\W', '') # replace with -> ''
df_price = df_price.astype(float,errors = 'ignore') # single -> error 
df_price = pd.to_numeric(df_price, errors='coerce') # .99 -> 99.
df_price = df_price.apply(lambda a : a/100) # scientific notation 
# shape 10841

df['Price'] = df_price # replacing old price with new price
```

```python
new_frame = df[['App','Category', 'Installs', 'Price']]
k = new_frame.groupby('Category').Installs.sum()  # number of installs

x = new_frame.Category.unique()
y = k.sort_values()
new_frame['Revenue'] = new_frame['Installs'] * new_frame['Price']
```

```python
plt.figure(figsize=(29,14))
plt.bar(x,y, color='green')
plt.title('Number of installs by Category')
plt.xticks(fontsize=14,rotation=60)
plt.yticks(fontsize=14.5)

           
plt.show()
```

```python
# unique category 
new_frame.Category.unique() 

k = new_frame.groupby('Category').Revenue.sum()
new_frame[(new_frame.Price > 0) & (new_frame.Category.isin(["COMICS", "HOUSE_AND_HOME","BEAUTY"]))] 

# no revenue in this COMICS", "HOUSE_AND_HOME" and "BEAUTY" category

x = new_frame.Category.unique()
y = k.sort_values()
y
```

```python
plt.figure(figsize=(29,14))
plt.bar(x,y, color='green')
plt.title('Revenue by Category')
plt.xticks(fontsize=14,rotation=60)
plt.yticks(fontsize=14.5)

           
plt.show()
```

```python
# top three based on paid

new_frame.groupby('Category').Revenue.sum().nlargest(3)
```

```python
# top three based on installs
new_frame.groupby('Category').Installs.sum().nlargest(3)
```
