```python
playstor_data = pd.read_csv('https://download1589.mediafire.com/4rfclscdaqhg/6zljdqj08o2sm30/googleplaystore.csv')
# link is changeable
```

```python
cols = ['App', 'Category', 'Rating', 'Reviews', 'Size', 'Installs', 'Type',
       'Price', 'Content Rating', 'Genres', 'Last Updated', 'Current Ver',
       'Android Ver']
```

```python
df = playstor_data.drop([10472]) # deleting the row
df.drop_duplicates(inplace=True) # removing duplictaes from all over the dataset
```

```python
(df
 [cols]
 .assign(Rating=df.Rating.fillna(0),
         Installs=df.Installs.str.replace(r'\W','',regex=True).astype(float),
         Price=df.Price.str.replace(r'\W','',regex=True).astype(float),
         Reviews=df.Reviews.astype(float),
         Type =df.Type.fillna(0),
         Content_Rating= df['Content Rating'].fillna('0'),
         Current_Ver=df['Current Ver'].fillna('0'),
         Android_Ver=df['Android Ver'].fillna('0')
        )
 .describe()
 .isna().sum()
)
```
